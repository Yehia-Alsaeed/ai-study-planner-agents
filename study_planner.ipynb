## Transformer Embedding Model

This loads a pre trained Transformer model (all-MiniLM-L6-v2) and uses its word embeddings to semantically classify subject difficulty. This component feeds into the student profiling phase.

```python
!pip install -q sentence-transformers scikit-learn

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
print("Transformer model loaded:", embedding_model)

difficulty_references = {
    "hard": [
        "neural networks", "transformers", "backpropagation",
        "attention mechanism", "deep learning", "natural language processing",
        "gradient descent", "embeddings"
    ],
    "medium": [
        "databases", "SQL queries", "network security",
        "encryption", "protocols", "data modeling"
    ],
    "easy": [
        "basic programming", "introduction to computing",
        "fundamentals", "overview", "history of computers"
    ]
}

ref_embeddings = {
    level: embedding_model.encode(examples)
    for level, examples in difficulty_references.items()
}

def classify_subject_difficulty(subject_name: str) -> str:
    """
    Uses Transformer word embeddings and cosine similarity
    to classify a subject as hard, medium, or easy.
    """
    subject_emb = embedding_model.encode([subject_name])
    scores = {}
    for level, ref_embs in ref_embeddings.items():
        similarities = cosine_similarity(subject_emb, ref_embs)
        scores[level] = float(np.max(similarities))
    predicted = max(scores, key=scores.get)
    return predicted, scores


print("\nEmbedding model ready for use in the profiling pipeline.")
```

```python
import os
import getpass
import json
from datetime import datetime

from crewai import LLM
from torch.utils.tensorboard import SummaryWriter

os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OPENAI_API_KEY: ")
os.environ["SERPER_API_KEY"] = getpass.getpass("Enter your SERPER_API_KEY: ")

llm_low = LLM(
    model="gpt-4o-mini",
    temperature=0.2,
    max_retries=3
)

llm_mid = LLM(
    model="gpt-4o-mini",
    temperature=0.6,
    max_retries=3
)

llm_high = LLM(
    model="gpt-4o-mini",
    temperature=0.8,
    max_retries=3
)


def save_variation_result(filename: str, result) -> str:
    """Save a CrewAI variation result as a text file in the Drive project folder."""
    path = os.path.join(RESULTS_DIR, filename)
    with open(path, "w", encoding="utf-8") as file:
        file.write(str(result))
    return path


print("Setup complete")
```

## Tools Setup

This initializes the external tools the agents will use

```python
from crewai_tools import SerperDevTool
from crewai.tools import tool

search_tool = SerperDevTool(n_results=3)

@tool("Calculator")
def calculator_tool(expression: str) -> str:
    """Evaluates a mathematical expression and returns the result as a string."""
    try:
        allowed = set("0123456789+-*/(). ")
        if not all(c in allowed for c in expression):
            return "Error: invalid characters in expression"
        return str(eval(expression))
    except Exception as e:
        return f"Error: {e}"

@tool("SubjectDifficultyClassifier")
def difficulty_classifier_tool(subject_name: str) -> str:
    """
    Uses a pre-trained Transformer model (all-MiniLM-L6-v2) and cosine similarity
    to classify a subject as hard, medium, or easy based on semantic similarity
    to reference examples.
    """
    predicted, scores = classify_subject_difficulty(subject_name)
    return (
        f"{subject_name} -> {predicted} "
        f"(hard: {scores['hard']:.3f}, medium: {scores['medium']:.3f}, easy: {scores['easy']:.3f})"
    )

print("All tools are ready")
```

## Agent Definitions

This defines the 4 agents that make up the multi-agent system

```python
from crewai import Agent

profiler_agent = Agent(
    role="Student Profiler",
    goal="Extract and structure all student information including subjects, difficulty levels, exam dates, and available study hours per day. Use the SubjectDifficultyClassifier tool to objectively classify the difficulty of each subject using semantic embeddings, then build a clean structured profile.",
    backstory="An expert academic advisor who uses AI-powered semantic tools to objectively assess subject difficulty before building a student profile. You never guess difficulty â€” you always call the SubjectDifficultyClassifier tool for each subject.",
    tools=[difficulty_classifier_tool],
    llm=llm_low,
    verbose=True,
)

generator_agent = Agent(
    role="Study Plan Generator",
    goal="Generate a realistic and balanced day-by-day study plan based on the student profile, using search to find effective study strategies per subject and the calculator to distribute hours fairly",
    backstory="An experienced academic coach who specializes in building personalized study schedules that are practical, balanced, and tailored to each student's strengths and deadlines",
    tools=[search_tool, calculator_tool],
    llm=llm_low,
    verbose=True,
)

critic_agent = Agent(
    role="Plan Critic",
    goal="Critically evaluate the study plan against the student's constraints, identify problems like overloaded days, neglected subjects, and missing buffer days before exams, and return a quality score out of 10 with specific issues listed",
    backstory="A strict and detail-oriented academic reviewer who carefully checks study plans for flaws, imbalances, and unrealistic expectations",
    tools=[calculator_tool],
    llm=llm_low,
    verbose=True,
)

optimizer_agent = Agent(
    role="Plan Optimizer",
    goal="Take the original plan and the critic's feedback and produce a final optimized study schedule that fixes all identified issues, with a short explanation of every change made",
    backstory="A senior academic strategist who takes critical feedback and transforms flawed study plans into polished, realistic, and highly effective schedules",
    tools=[search_tool, calculator_tool],
    llm=llm_low,
    verbose=True,
)

print("All 4 agents are ready.")
```

## Task Definitions and Crew Assembly

This defines what each agent must do and wires everything together into a single crew

```python
from crewai import Task, Crew, Process

profiling_task = Task(
    description=(
        "Take the following raw student input and extract all relevant information. "
        "For EVERY subject mentioned, call the SubjectDifficultyClassifier tool to determine its difficulty â€” "
        "do not skip any subject and do not guess difficulty yourself. "
        "Then produce a clean structured profile including: each subject with its classified difficulty, "
        "the exam day per subject, the total number of study days, and the daily available study hours. "
        "IMPORTANT: Count study days inclusively â€” if the first day is Day 1 and the last exam is on Day N, "
        "there are exactly N total study days (Day 1 to Day N inclusive = N days, not N-1). "
        "Raw input: {student_input}"
    ),
    expected_output=(
        "A clearly structured student profile with: "
        "(1) each subject with its classifier-determined difficulty level, "
        "(2) exam day per subject, "
        "(3) total study days available (counted inclusively), "
        "(4) daily available study hours."
    ),
    agent=profiler_agent,
)

generation_task = Task(
    description="""You MUST follow these steps in order before writing the schedule.

Step 1: For every subject in the student profile, use the search tool to search for
'best study strategies for [subject name] computer science university exam'.
Do this once per subject â€” for all subjects, no skipping.

Step 2: Use the calculator for exactly these three expressions, then stop â€” do not use the calculator again:
- total_hours = total_days * daily_hours
- hard_pool = total_hours * 0.6
- medium_pool = total_hours * 0.4

Step 3: Write the LOCKED DAY TABLE before anything else. Use exactly this format:
  [Subject with earliest exam]: Buffer = Day (exam_day - 1), Exam = Day exam_day
  [Subject with 2nd exam]:      Buffer = Day (exam_day - 1), Exam = Day exam_day
  [Subject with 3rd exam]:      Buffer = Day (exam_day - 1), Exam = Day exam_day
  [Subject with latest exam]:   Buffer = Day (exam_day - 1), Exam = Day exam_day
Buffer day = the day immediately before the exam day. Exam day = the day the exam occurs.

Step 4: Fill all remaining days (days NOT in the locked table) with new study material.
Rules â€” no exceptions:
- On a BUFFER day: label it "[Subject] BUFFER DAY â€” review only, no new material for [subject]"
- On an EXAM day: label it "[Subject] EXAM DAY â€” light review only"
- No day can exceed the student's daily available hours
- Every day from Day 1 to the last exam day must be assigned something
- Hard subjects get proportionally more study days than medium subjects
- Never schedule new material for a subject on its buffer day or exam day
- Never schedule any study for a subject after its exam day has passed

Student profile: {student_input}""",
    expected_output=(
        "First: the locked day table. "
        "Then: a full day-by-day study plan from Day 1 to the last exam day. "
        "Every buffer day and exam day must be clearly labeled for every subject."
    ),
    agent=generator_agent,
)

critique_task = Task(
    description=(
        "Carefully review the generated study plan against the student profile. "
        "Check every one of the following points:\n"
        "1. Does any study day exceed the daily available hours limit from the profile?\n"
        "2. Is any subject scheduled on a day AFTER its exam day has already passed?\n"
        "3. Does every exam have a buffer day on the day IMMEDIATELY before it "
        "(buffer day = exam_day - 1)?\n"
        "4. Is every day from Day 1 to the last exam day covered â€” no gaps?\n"
        "5. Do hard subjects have more total study hours than medium subjects?\n"
        "6. Does EVERY subject have its own EXAM DAY label â€” not BUFFER DAY â€” on its correct exam day?\n\n"
        "For each subject, count its study hours by reading the plan directly. "
        "Reference only the student profile for all exam dates and constraints."
    ),
    expected_output=(
        "A numbered list of every issue found, with specific day numbers and subject names cited. "
        "End with exactly: Quality Score: X/10"
    ),
    agent=critic_agent,
)

optimization_task = Task(
    description=(
        "Using the original study plan and every issue the critic identified, produce a corrected final schedule.\n\n"
        "STEP 1 â€” Write this verification table before anything else:\n"
        "  [Subject 1]: exam = Day X  â†’  buffer = Day X-1  (buffer is always exam_day MINUS ONE)\n"
        "  [Subject 2]: exam = Day Y  â†’  buffer = Day Y-1\n"
        "  [Subject 3]: exam = Day Z  â†’  buffer = Day Z-1\n"
        "  [Subject 4]: exam = Day W  â†’  buffer = Day W-1\n\n"
        "CRITICAL EXAMPLE â€” read carefully:\n"
        "  If Security exam = Day 10, then Security buffer = Day 9 (NOT Day 10).\n"
        "  Day 9 must be labeled 'Security BUFFER DAY â€” review only'.\n"
        "  Day 10 must be labeled 'Security EXAM DAY â€” light review only'.\n"
        "  Day 10 is the EXAM, never the buffer.\n\n"
        "STEP 2 â€” Write the final plan only after completing Step 1. Enforce these rules:\n"
        "1. Every subject's buffer day label = '[Subject] BUFFER DAY' on day exam_day - 1.\n"
        "2. Every subject's exam day label = '[Subject] EXAM DAY' on its exact exam day.\n"
        "3. Never schedule new study for a subject after its exam day. "
        "Example: if Deep Learning exam = Day 12, then Day 13 must NOT contain Deep Learning.\n"
        "4. No day may exceed the student's daily available hours.\n"
        "5. All values â€” subject names, exam days, hour limits â€” must come from the student profile."
    ),
    expected_output=(
        "First: the verification table with all four subjects. "
        "Then: the final optimized day-by-day study plan in clean plain text. "
        "Then: a bullet list of every change made and the reason. "
        "End with exactly: Quality Score: X/10"
    ),
    agent=optimizer_agent,
)

study_crew = Crew(
    agents=[profiler_agent, generator_agent, critic_agent, optimizer_agent],
    tasks=[profiling_task, generation_task, critique_task, optimization_task],
    process=Process.sequential,
    verbose=True,
)

print("The crew is assembled and ready to run.")
```

## Running 3 Variations

This pruns the full crew 3 times with different temperature settings and stores results for comparison

```python
student_input = "I have 4 subjects: NLP, Deep Learning, Security, and Database. My exams are: NLP on day 14, Deep Learning on day 12, Security on day 10, and Database on day 8. I have 4 hours available to study every day starting from today which is day 1."

profiler_agent.llm = llm_low
generator_agent.llm = llm_low
critic_agent.llm = llm_low
optimizer_agent.llm = llm_low


study_crew = Crew(
    agents=[profiler_agent, generator_agent, critic_agent, optimizer_agent],
    tasks=[profiling_task, generation_task, critique_task, optimization_task],
    process=Process.sequential,
    verbose=True,
)

result_low_temp = study_crew.kickoff(inputs={"student_input": student_input})
print("VARIATION 1 â€” Temperature 0.2")
print(result_low_temp)

result_low_temp_path = save_variation_result("variation_a_temperature_0_2.txt", result_low_temp)
print(f"Variation A saved to: {result_low_temp_path}")
```

## Evaluation and TensorBoard Logging

This evaluates the variations using defined metrics and logs the results to TensorBoard for comparison

```python
import re

def plan_completeness_score(result, subjects):
    result_text = str(result).lower()
    matches = sum(1 for subject in subjects if subject.lower() in result_text)
    return matches / len(subjects)

def constraint_satisfaction_score(result):
    hour_mentions = re.findall(r"(\d+(?:\.\d+)?)\s*hours?", str(result), re.IGNORECASE)
    hours = [float(value) for value in hour_mentions]
    if not hours:
        return 1.0
    valid_hours = sum(1 for value in hours if value <= 4)
    return valid_hours / len(hours)

def quality_score(result):
    match = re.search(r"Quality Score:\s*(\d+(?:\.\d+)?)\s*/\s*10", str(result), re.IGNORECASE)
    if not match:
        return 0.5
    return float(match.group(1)) / 10

subjects = ["NLP", "Deep", "Security", "Database"]

scores_low = {
    "completeness": plan_completeness_score(result_low_temp, subjects),
    "constraint": constraint_satisfaction_score(result_low_temp),
    "quality": quality_score(result_low_temp),
}

scores_mid = {
    "completeness": plan_completeness_score(result_mid_temp, subjects),
    "constraint": constraint_satisfaction_score(result_mid_temp),
    "quality": quality_score(result_mid_temp),
}

scores_high = {
    "completeness": plan_completeness_score(result_high_temp, subjects),
    "constraint": constraint_satisfaction_score(result_high_temp),
    "quality": quality_score(result_high_temp),
}

print("Scores for Temperature 0.2:")
print(scores_low)
print("Scores for Temperature 0.6:")
print(scores_mid)
print("Scores for Temperature 0.8:")
print(scores_high)

all_scores = {
    "temperature_0_2": scores_low,
    "temperature_0_6": scores_mid,
    "temperature_0_8": scores_high,
}

scores_path = os.path.join(RESULTS_DIR, "variation_scores.json")
with open(scores_path, "w", encoding="utf-8") as file:
    json.dump(all_scores, file, indent=2)

print(f"Scores saved to: {scores_path}")
```

```python
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter(log_dir=LOG_DIR)

writer.add_scalar("Metrics/Plan_Completeness", scores_low["completeness"], 0)
writer.add_scalar("Metrics/Plan_Completeness", scores_mid["completeness"], 1)
writer.add_scalar("Metrics/Plan_Completeness", scores_high["completeness"], 2)
writer.add_scalar("Metrics/Constraint_Satisfaction", scores_low["constraint"], 0)
writer.add_scalar("Metrics/Constraint_Satisfaction", scores_mid["constraint"], 1)
writer.add_scalar("Metrics/Constraint_Satisfaction", scores_high["constraint"], 2)
writer.add_scalar("Metrics/Quality_Score", scores_low["quality"], 0)
writer.add_scalar("Metrics/Quality_Score", scores_mid["quality"], 1)
writer.add_scalar("Metrics/Quality_Score", scores_high["quality"], 2)

writer.close()

metrics_table = (
    "Metric                  | Temp 0.2 | Temp 0.6 | Temp 0.8\n"
    "------------------------|----------|----------|--------\n"
    f"Plan Completeness       |   {scores_low['completeness']:.2f}   |   {scores_mid['completeness']:.2f}   |  {scores_high['completeness']:.2f}\n"
    f"Constraint Satisfaction |   {scores_low['constraint']:.2f}   |   {scores_mid['constraint']:.2f}   |  {scores_high['constraint']:.2f}\n"
    f"Quality Score           |   {scores_low['quality']:.2f}   |   {scores_mid['quality']:.2f}   |  {scores_high['quality']:.2f}"
)

print(metrics_table)

metrics_table_path = os.path.join(RESULTS_DIR, "metrics_table.txt")
with open(metrics_table_path, "w", encoding="utf-8") as file:
    file.write(metrics_table)

print(f"Metrics table saved to: {metrics_table_path}")
print(f"TensorBoard logs saved to: {LOG_DIR}")

%load_ext tensorboard
%tensorboard --logdir "$LOG_DIR"
```
